Step 3.a:

 Para tener una mejor noci√≥n de cada flag en nuestra tabla decidimos hacer un descriptivo auxiliar de cada concepto ya que al momento de hacer un count podamos ver que especificamente estamos contando y no tengamos que pensar que es 0/1

```python
#descripci√≥n de si es owner del veh√≠culo o no
df['VEHICLE_OWNERSHIP_desc'] = ['Owner' if x == 1 else 'Non_Owner' for x in df['VEHICLE_OWNERSHIP']]
```

```python
#descripci√≥n de si est√° casado o no
df['MARRIED_desc'] = ['Married' if x == 1 else 'Non_Married' for x in df['MARRIED']]
```

```python
#descripci√≥n de si tiene hijos o no
df['CHILDREN_desc'] = ['With_CHILDREN' if x == 1 else 'WO_CHILDREN' for x in df['CHILDREN']]
```

```python
#descripci√≥n de si tuvo o no algun siniestro
df['OUTCOME_desc'] = ['Claim' if x == 1 else 'WO_Claim' for x in df['OUTCOME']]
```

Step 3.b:

Rango para credit score

```python
#vamos a generar un rango de score para credit score
score_a = 0.4
score_b = 0.7
df['CREDIT_SCORE_range'] = ''
```

```python
df.loc[df['CREDIT_SCORE'] < score_a, 'CREDIT_SCORE_range' ] = 'RIESGO_ALTO'
df.loc[np.logical_and(df['CREDIT_SCORE'] >= score_a, df['CREDIT_SCORE'] < score_b) , 'CREDIT_SCORE_range' ] = 'RIESGO_MEDIO'
df.loc[df['CREDIT_SCORE'] >= score_b, 'CREDIT_SCORE_range'] = 'RIESGO_BAJO'
```

###  ROC Curve gr√°fico y m√©trica

Es una herramienta gr√°fica utilizada para evaluar el rendimiento de un modelo de clasificaci√≥n en diferentes umbrales de decisi√≥n. La curva ROC muestra la relaci√≥n entre la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR) a medida que se ajusta el umbral de decisi√≥n del modelo.

El √°rea bajo la curva ROC (AUC-ROC) es otra m√©trica importante asociada con la curva ROC. Representa el √°rea debajo de la curva y proporciona una medida de cu√°n bien el modelo puede distinguir entre las clases. Un valor de AUC-ROC cercano a 1 indica un mejor rendimiento del modelo en la clasificaci√≥n, que es lo que vamos a calcular:

```python
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(predictions_modelo1_test, y_test)
print('% de √°rea debajo de curva roc sobre el set de test :', roc_auc)
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="left"><font color=white>‚òëÔ∏è Highlights de ROC Curve area:
</font></h3>

Ya que tenemos variables cualitativas (dicotomicas) en un an√°lisis PCA, primero vamos a convertir las variables num√©ricas utilizando t√©cnicas adecuadas, como la codificaci√≥n de variables dummy.
No agrupamos sino reducimos dimensionalidad:
OBJETIVO: Encontrar nuevas dimensiones (llamadas componentes principales) que son combinaciones lineales de las dimensiones originales. Estas nuevas dimensiones est√°n ordenadas de tal manera que la primera componente principal captura la m√°xima varianza en los datos, la segunda componente principal captura la siguiente m√°xima varianza, y as√≠ sucesivamente. Entonces buscamos  simplificar datos complejos y mejorar la comprensi√≥n y la eficiencia en el an√°lisis de datos.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

Vamos a entrenar nuestro modelo de RandomForest con el objetivo de evaluar el rendimiento de un modelo y reducir el riesgo de sobreajuste.

## Statified K - fold (con 5 divisiones)


Usaremos Stratified K-Fold, el cual es una variante de la validaci√≥n cruzada K-Fold que mantiene la proporci√≥n de las clases en cada fold. Se asegura de que cada fold tenga una distribuci√≥n de clases similar a la del conjunto de datos completo.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold, KFold
from sklearn.metrics import accuracy_score


# Definir los par√°metros que deseas probar en el GridSearch
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10]
}

# Crear el clasificador RandomForest
clf = RandomForestClassifier()

# Crear el objeto StratifiedKFold con 5 divisiones (k=5)
S_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Crear el objeto GridSearchCV
grid_search = GridSearchCV(clf, param_grid, cv=S_kfold, scoring='accuracy', return_train_score=True, verbose=3)

# Ajustar el GridSearch en los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Obtener los mejores par√°metros y el mejor estimador del GridSearch
best_params = grid_search.best_params_


print("Mejores hiperpar√°metros:", best_params)


```

```python
# Realizar predicciones en todo el conjunto de datos utilizando el mejor estimador

best_estimator = grid_search.best_estimator_
y_pred = best_estimator.predict(X_train)

# Calcular la precisi√≥n
accuracy = accuracy_score(y_train, y_pred)
accuracy
```

Luego de las iteraciones, lo que nos est√° diciendo en conclusi√≥n la t√©cnica de Stratified K-fold que los mejores hiperpar√°metros son {'max_depth': 10, 'n_estimators': 100} mi profundidad es 10 y cantidad de √°rboles son 100.
Con estos mejores estimadores hacemos la predicci√≥n el modelo de RandomForest, es decir reentrenamos el modelo con dichos hiperpar√°metros.

Procedemos a calcular la accuracy= 86%

Podemos ver todos los resultados de nuestro grid search:

```python
# Obtener los scores de todas las combinaciones de par√°metros
results = grid_search.cv_results_
results
```

Todos esos resultados de Grid search, podemos dejarlos en un Dataframe:

```python
# guardemos lo m√°s importante dentro de un dataframe
df = pd.DataFrame(results)


df = df[['mean_train_score', 'mean_test_score', 'params']]
df.round(3).sort_values("mean_test_score", ascending=False)
```

De los 4 splits y calcula la media de cada uno, se puede ver en cada uno las mejores combinaciones de cada uno. De este modo, se puede ver que la t√©cnica de cross validation busca siempre que la varianza sea bastante baja y el bias tambi√©n. Siendo el objetivo principal no tengamos overfitting, es con una profundidad de 10 y 100 √°rboles


## K - fold (con 5 divisiones)


Usaremos K-Fold, el cual consiste en dividir el conjunto de datos en K partes (folds) y realizar K iteraciones del proceso de entrenamiento y evaluaci√≥n. En cada iteraci√≥n, un fold se utiliza como conjunto de prueba, mientras que los K-1 folds restantes se utilizan para entrenar el modelo. Este proceso se repite K veces, cada vez utilizando un fold diferente como conjunto de prueba.

```python
# Crear el objeto KFold con 5 divisiones (k=5)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Crear el objeto GridSearchCV con KFold
grid_search = GridSearchCV(clf, param_grid,  refit=True, scoring='accuracy', return_train_score=True)

# Ajustar el GridSearch en los datos de entrenamiento
grid_search.fit(X_train, y_train)

# Obtener los mejores par√°metros y el mejor estimador del GridSearch
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

# Imprimir los mejores par√°metros encontrados
print("Mejores par√°metros:", best_params)
```

K fold: a diferencia de stratified K fold la distribucion de las clases, no nos aseguramos que las clases esten distribuidas proporcionalmente. Vemos en esta t√©cnica la conclusi√≥n es distinta:
los mejores hiperpar√°metros son {'max_depth': 10, 'n_estimators': 200} mi profundidad es 10 y cantidad de √°rboles son 200 (antes 100).

```python
# Realizar predicciones utilizando el mejor estimador
y_pred = best_estimator.predict(X_train)

# Calcular la precisi√≥n
accuracy = accuracy_score(y_train, y_pred)
print("Precisi√≥n:", accuracy)
```
